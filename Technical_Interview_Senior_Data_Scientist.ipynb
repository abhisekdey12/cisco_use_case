## Technical Interview - Senior Data Scientist 

### Introduction 


We would like you to forecast the next 3 months of demand for one of our business entities, and present your findings. 

The attached file (`bookings_data.parquet`) contains bookings data for this entity, as well as bookings data broken out into the 3 different subsegments of this entity. There is also Lead Time information for delivery of products in each segment (actual lead times measured, lead times predicted by Cisco, and lead times requested by clients), as well as an array of external macroeconomic features that we might expect to be predictive of demand.

Cisco's internal features have all been normalised, and the `net_bookings_3m | *` columns contain **monthly time series of rolling 3 months of bookings**. For example the value at date `'2023-12-31'` contains the cumulative bookings for October, November, and December of 2023. 

The relationship between these columns is such that:

`net_bookings_3m | Entity` == `net_bookings_3m | Segment_A`+`net_bookings_3m | Segment_B`+`net_bookings_3m | Segment_C`
